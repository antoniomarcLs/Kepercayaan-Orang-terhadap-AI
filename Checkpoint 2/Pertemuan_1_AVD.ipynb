### **Outliers Values**

This point explain an outliers in some column that has integer or float64 as his data type in purpose to get good and accurate insight.

```python
results = []

cols = df.select_dtypes(include=['float64', 'int64'])

for col in cols:
  q1 = df[col].quantile(0.25)
  q3 = df[col].quantile(0.75)
  iqr = q3 - q1
  lower_bound = q1 - 1.5*iqr
  upper_bound = q3 + 1.5*iqr
  outliers = df[(df[col] < lower_bound) | (df[col] > upper_bound)]
  percent_outliers = (len(outliers)/len(df))*100
  results.append({'Column': col, 'Outliers percentage': percent_outliers})

results_df = pd.DataFrame(results)
results_df.set_index('Column', inplace=True)
results_df = results_df.rename_axis(None, axis=0).rename_axis('Colum', axis=1)

display(results_df)
```

As we can see from the information the outliers percentage is so low like

* ai_confidence_percentage, has no outliers

* response_character_count, has no outliers

* trust_score_out_of_10, has an outliers of (0.3%)

* verifivation_duration_mins, has no outliers

* answer_accuracy_percentage, has an outliers of (0.3%)

# **Data Explorations (EDA)**

## **Comparison**

Activity: Comparing the confidence to answering of each AI agent and place them side by side

Purpose: To see which one of the AI agent has the biggest confidence to answer a question among all of them

Visualization: Bar chart

```python
ai_model_name = df.groupby('ai_model_name')['response_character_count'].sum().sort_values(ascending=False)

plt.figure(figsize=(10,6))
sns.barplot(x=ai_model_name.index, y=ai_model_name.values, palette='viridis', hue=ai_model_name.index, legend=False)
plt.title('AI confidence of each AI agent')
plt.xlabel('ai_model_name')
plt.ylabel('ai_confidence_percentage')
plt.xticks(rotation=45)
plt.show()
```

## **Composition**

Activity: see the proportion the the data and to see their own value with other by percentage

Purpose: Knowing which one of the AI agent that has the biggest confidence on their answer

Visualization: Pie chart

```python
urgency_level = df.groupby('ai_model_name')['ai_confidence_percentage'].sum().sort_values(ascending=False)
urgency_level_top3 = urgency_level.head(3)

plt.figure(figsize=(10, 8))
urgency_level_top3.plot(kind='pie', autopct='%1.1f%%', startangle=140, colors=plt.cm.Paired.colors)
plt.title('Top 3 AI model with highest confidence percentage')
plt.ylabel('urgency_level')
plt.axis('equal')
plt.show()
```

## **Distribution**

Activity: Exploring the value of the answer accurracy

Purpose: Knowing the frequency of the answer accurracy data are most of them have big value or not and to know are there any skewness at the data

Visualization: Histogram

```python
plt.figure(figsize=(10, 6))
sns.histplot(df['answer_accuracy_percentage'], bins=20, kde=True)
plt.title('Distribution of answer accurracy percentage')
plt.xlabel('answer accurracy')
plt.ylabel('Frekuensi')
plt.show()
```

## **Relationship**

Activity: analyze a corrulation between AI confidence, answer accuracy, verification times, response character count, trust score out of 10

Purpose: To know about their relation with each other and how much impact they have to each other, are they have directly proportional or not or maybe they have no relationship at all

Visualization: Heatmap

```python
plt.figure(figsize=(8, 6))
sns.heatmap(data=df[['ai_confidence_percentage', 'answer_accuracy_percentage', 'verification_duration_mins', 'response_character_count', 'trust_score_out_of_10']].corr(),
            annot=True,
            cmap='viridis',
            fmt='.2f')
plt.title('Korelasi antara Sales dan Profit')
plt.show()
```
